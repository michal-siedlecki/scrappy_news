```python
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
# from nltk.stem import WordNetLemmatizer
from stempel import StempelStemmer
from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
from nltk.corpus import wordnet as wn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import model_selection, naive_bayes, svm
from sklearn.metrics import accuracy_score
import pickle

f = open('finalized_model.sav', 'rb')
model = pickle.load(f)
f.close()

sport = pd.read_csv(r"headers_corp.csv",encoding='utf-8').iloc[1000:1010]
nonsport = pd.read_csv(r"headers_corp.csv",encoding='utf-8').iloc[-110:-100]
Corpus = pd.concat([sport,nonsport]).reset_index(drop=True)

samples = [x for x in Corpus['text']]
print(samples)
```

```python
samples = [word_tokenize(entry, language='polish') for entry in samples]
print(len(samples))
```

```python
Tfidf_vect = TfidfVectorizer(max_features=1000)
samplesSeries = pd.Series(samples)
Tfidf_vect.fit(samplesSeries)
```

```python
my_test = Tfidf_vect.transform(samples)
print(my_test)
print(type(my_test))

predictions = model.predict(my_test)

for tup in zip(samples,predictions):
    print(tup[0])
    print(tup[1])
```
