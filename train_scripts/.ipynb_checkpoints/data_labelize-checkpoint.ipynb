```python
# Label training data and save it as csv
import pandas as pd
import numpy as np
import os

headers_amounts = []
headers_files = [x for x in os.listdir(r'E:\ms-prywatne\scrappy_news\train_data') if x.endswith('.txt')]
```

```python
def countHeaders(file):
    with open(rf'E:\ms-prywatne\scrappy_news\train_data\{file}', 'r', encoding='utf-8') as f:
        headers_amounts.append(len(f.readlines()))



def saveToCsv(file,labelNum):
    with open(rf'E:\ms-prywatne\scrappy_news\train_data\{file}', 'r', encoding='utf-8') as f:
        headers = f.readlines()[:min(headers_amounts)]

    with open(r'E:\ms-prywatne\scrappy_news\train_data\corpus.csv', 'a', encoding='utf-8') as f:
        f.write('text,label')
        for header in headers:
            f.write(header.strip().replace(',','')+f',__label__{labelNum}\n')
```

```python
for file in headers_files:
    countHeaders(file)
```

```python
for num, file in enumerate(headers_files):
    saveToCsv(file, num+1)
```
