### onet politico scrapper

```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

browser = webdriver.Chrome()

browser.get("https://wiadomosci.onet.pl/politico")
time.sleep(5)

elem = browser.find_element_by_tag_name("body")

pages = 300

while pages:
    elem.send_keys(Keys.END)
    time.sleep(0.8)
    pages -= 1

onet_headers = browser.find_elements_by_xpath("//div[@class='boxTitle']//h3[@class='title']")

with open('politics.txt', 'w') as f:
    for x in onet_headers:
        f.write(x.text+"\n")
```

### tvn scrapper

```python
import time
import requests
from flask import Flask, render_template
from bs4 import BeautifulSoup

num = 2
tvn_headers = []

for _ in range(90):
    url = 'https://www.tvn24.pl/wiadomosci-ze-swiata,2/{}'.format(num)
    time.sleep(0.2)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    small_headers = soup.find_all(class_='size28')
    large_headers = soup.find_all(class_='size36')
    xlarge_headers = soup.find_all(class_='size36')
    for header in small_headers:
        tvn_headers.append(header.text)
    for header in large_headers:
        tvn_headers.append(header.text)
    for header in xlarge_headers:
        tvn_headers.append(header.text)
    num +=1

tvn_headers = list(set(tvn_headers))

with open('politics.txt', 'a+') as f:
    for x in tvn_headers:
        f.write(x+"\n")
```

### wpolityce scrapper

```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

browser = webdriver.Chrome()

browser.get("https://wpolityce.pl/polityka")
time.sleep(5)
elem = browser.find_element_by_tag_name("body")
pages = 50
progress = '*'
while pages:
    elem.send_keys(Keys.END)
    time.sleep(0.8)
    pages -= 1
    if pages%10 == 0:
        print('progress '+progress)
        progress += '*'

wpolityce_headers = browser.find_elements_by_xpath("//h3[@class='ellipsis']")
wpolityce_headers = list(set([x.text for x in wpolityce_headers]))
print(len(wpolityce_headers))
```

```python
n = [x.split("\n") for x in wpolityce_headers]
wpolityce_headers = []
for l in n:
    for elem in l:
        wpolityce_headers.append(elem)
        
for x in wpolityce_headers[:10]:
    print(x)

wpolityce_headers = list(set([x for x in wpolityce_headers]))
```

```python
with open('politics.txt', 'a+') as f:
    for x in wpolityce_headers:
        f.write(x+"\n")
```

### spider's web scrapper

```python
import time
import requests
from flask import Flask, render_template
from bs4 import BeautifulSoup

num = 2
spiders_headers = []

for _ in range(90):
    url = 'https://www.spidersweb.pl/page/{}'.format(num)
    time.sleep(0.2)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    small_headers = soup.find_all(class_='postlink-inner')
    for header in small_headers:
        spiders_headers.append(header.text)
    
    num +=1
    
len(spiders_headers)
```

```python
for x in spiders_headers:
    print(x)
```

```python
with open('tech.txt', 'w', encoding='utf-8') as f:
    
    for j in spiders_headers:
        f.write(j+"\n")
```

### computerworld

```python
import time
import requests
from bs4 import BeautifulSoup

num = 2
cw_headers = []

for _ in range(25):
    url = 'https://www.computerworld.pl/temat/technologie/{}.html'.format(num)
    time.sleep(0.2)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    headers = soup.find_all(class_='title')
    for header in headers:
        cw_headers.append(header.text)
    
    num +=1
print(len(cw_headers))
```

```python
def save_headers(headers_list, filename):
    with open(filename, 'a', encoding='utf-8') as f:
        for header in headers_list:
            f.write(header+"\n")
```

```python
def getHeadersLen(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        print(len(f.readlines()))
```

### portal technologiczny

```python
import time
import requests
from bs4 import BeautifulSoup

page_num = 2
pt_headers = []

for _ in range(41):
    time.sleep(1)
    url = f'https://portaltechnologiczny.pl/category/technews/page/{page_num}'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    content = soup.find_all('header',class_='genpost-entry-header')
    for div in content:
        pt_headers.append(div.a.text)
        
    page_num += 1

print(len(pt_headers))
```

```python
save_headers(pt_headers, 'tech.txt')
```

```python
getHeadersLen('tech.txt')
```

### super express

```python
import time
import requests
from bs4 import BeautifulSoup

page_num = 2
se_headers = []

for _ in range(58):
    time.sleep(0.3)
    url = f'https://www.se.pl/temat/prognoza-pogody,tg-FFep-HXLa-Veo4/?page={page_num}'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    content = soup.find_all('div',class_='element__headline')
    for div in content[:-4]:
        se_headers.append(div.a.text.strip())

    page_num += 1
print(len(se_headers))
```

```python
save_headers(se_headers, 'weather.txt')
```

### celebrities news

```python
import time
import requests
from bs4 import BeautifulSoup

page_num = 2
plotek_headers = []

for _ in range(200):
    time.sleep(0.3)
    url = f'http://www.plotek.pl/plotek/0,78649.html?str={page_num}_25423735'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    content = soup.find_all('article',class_='article')
    for article in content:
        plotek_headers.append(article.header.h2.a.text)
    page_num += 1
    

print(len(plotek_headers))
```

```python
save_headers(plotek_headers[:1000], 'rumors.txt')
```
