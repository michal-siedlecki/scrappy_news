### onet politico scrapper

```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

browser = webdriver.Chrome()

browser.get("https://wiadomosci.onet.pl/politico")
time.sleep(5)

elem = browser.find_element_by_tag_name("body")

pages = 300

while pages:
    elem.send_keys(Keys.END)
    time.sleep(0.8)
    pages -= 1

onet_headers = browser.find_elements_by_xpath("//div[@class='boxTitle']//h3[@class='title']")

with open('politics.txt', 'w') as f:
    for x in onet_headers:
        f.write(x.text+"\n")
```

### tvn scrapper

```python
import time
import requests
from flask import Flask, render_template
from bs4 import BeautifulSoup

num = 2
tvn_headers = []

for _ in range(90):
    url = 'https://www.tvn24.pl/wiadomosci-ze-swiata,2/{}'.format(num)
    time.sleep(0.2)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    small_headers = soup.find_all(class_='size28')
    large_headers = soup.find_all(class_='size36')
    xlarge_headers = soup.find_all(class_='size36')
    for header in small_headers:
        tvn_headers.append(header.text)
    for header in large_headers:
        tvn_headers.append(header.text)
    for header in xlarge_headers:
        tvn_headers.append(header.text)
    num +=1

tvn_headers = list(set(tvn_headers))

with open('politics.txt', 'a+') as f:
    for x in tvn_headers:
        f.write(x+"\n")
```

### wpolityce scrapper

```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

browser = webdriver.Chrome()

browser.get("https://wpolityce.pl/polityka")
time.sleep(5)
elem = browser.find_element_by_tag_name("body")
pages = 50
progress = '*'
while pages:
    elem.send_keys(Keys.END)
    time.sleep(0.8)
    pages -= 1
    if pages%10 == 0:
        print('progress '+progress)
        progress += '*'

wpolityce_headers = browser.find_elements_by_xpath("//h3[@class='ellipsis']")
wpolityce_headers = list(set([x.text for x in wpolityce_headers]))
print(len(wpolityce_headers))
```

```python
n = [x.split("\n") for x in wpolityce_headers]
wpolityce_headers = []
for l in n:
    for elem in l:
        wpolityce_headers.append(elem)
        
for x in wpolityce_headers[:10]:
    print(x)

wpolityce_headers = list(set([x for x in wpolityce_headers]))
```

```python
with open('politics.txt', 'a+') as f:
    for x in wpolityce_headers:
        f.write(x+"\n")
```

### spider's web scrapper

```python
import time
import requests
from flask import Flask, render_template
from bs4 import BeautifulSoup

num = 2
spiders_headers = []

for _ in range(90):
    url = 'https://www.spidersweb.pl/page/{}'.format(num)
    time.sleep(0.2)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    small_headers = soup.find_all(class_='postlink-inner')
    for header in small_headers:
        spiders_headers.append(header.text)
    
    num +=1
    
len(spiders_headers)
```

```python
with open('tech.txt', 'w') as f:
    for x in spiders_headers:
        f.write(x+"\n")
```
