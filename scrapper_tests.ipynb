```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

news_num = 100
```

```python
class Feed:
    def __init__(self, name, category, url):
        self.name = name
        self.category = category
        self.url = url


    def getContent(self):
        page = requests.get(self.url)
        soup = BeautifulSoup(page.content, 'html.parser')
        return soup
```

```python
class Onet(Feed):
    def getPolitics(self):
        soup = self.getContent()
        divs = soup.find_all(class_="boxTitle")
        headers_list = [x.text for x in divs]
        return headers_list
    
    def getWeather(self):
        soup = self.getContent()
        divs = soup.find_all(class_="titleHolder")
        headers_list = [x.text for x in divs]
        return headers_list
        
                
```
