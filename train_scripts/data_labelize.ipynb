```python
# Label training data and save it as csv
import pandas as pd
import numpy as np

numlist = []

def countHeaders():
    with open('train_data/sport.txt', 'r', encoding='utf-8') as f:
        numlist.append(len(f.readlines()))
    with open('train_data/politics.txt', 'r', encoding='utf-8') as f:
        numlist.append(len(f.readlines()))
        

def sportCSV():
    with open('train_data/sport.txt', 'r', encoding='utf-8') as f:
        headers = f.readlines()[:min(numlist)]

    with open('train_data/corpus.csv', 'w', encoding='utf-8') as f:
        f.write('text,label')
        for header in headers:
            f.write(header.strip().replace(',','')+',__label__1\n')
            
            
def politicsCSV():
    with open('train_data/politics.txt', 'r', encoding='utf-8') as f:
        headers = f.readlines()[:min(numlist)]

    with open('train_data/corpus.csv', 'a+', encoding='utf-8') as f:
        f.write('text,label')
        for header in headers:
            f.write(header.strip().replace(',','')+',__label__2\n')
            

countHeaders()
sportCSV()
politicsCSV()
```
